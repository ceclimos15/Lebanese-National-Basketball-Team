```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

flb = pd.read_csv('la3ibe.csv') 
flb
```

```python
#flb['Height'] = pd.to_numeric(flb['Height'], errors='coerce')
#flb['FG%'] = pd.to_numeric(flb['FG%'], errors='coerce')
flb['Pts'] = pd.to_numeric(flb['Pts'], errors='coerce')
flb['Reb'] = pd.to_numeric(flb['Reb'], errors='coerce')
flb['Ast'] = pd.to_numeric(flb['Ast'], errors='coerce')
flb['Reb.1'] = pd.to_numeric(flb['Reb.1'], errors='coerce')
flb['Ast.1'] = pd.to_numeric(flb['Ast.1'], errors='coerce')
flb['Stl'] = pd.to_numeric(flb['Stl'], errors='coerce')
flb['Fo'] = pd.to_numeric(flb['Fo'], errors='coerce')
flb['Pts.1'] = pd.to_numeric(flb['Pts.1'], errors='coerce')
flb['FT%'] = pd.to_numeric(flb['FT%'], errors='coerce')
flb['Blk'] = pd.to_numeric(flb['Blk'], errors='coerce')
flb['Eff'] = pd.to_numeric(flb['Eff'], errors='coerce')
flb['To'] = pd.to_numeric(flb['To'], errors='coerce')
```

```python
flb.dtypes
```

```python
#convert the columnn 3P% to float
def convert_percent(value):
    if pd.isna(value) or value == '-':
        return np.nan
    else:
        value = value.replace(',', '.')  
        return float(value.rstrip('%')) 

flb['3P%'] = flb['3P%'].apply(convert_percent)
flb['FG%'] = flb['FG%'].apply(convert_percent)
flb[['3P%','FG%'] ]
```

```python
#convert the column 'height' to integer 
def convert_height(height_str):
        if (pd.isna(height_str)) or height_str == '-':
            return np.nan
        else :
            height_value = float(height_str.replace('m',''))  # Use float if decimal places are expected
            return height_value    
    
# Apply the conversion function to 'Height' column
flb['Height'] = flb['Height'].apply(convert_height)
```

```python
new_flb = flb[['Player', 'Height','MIN','3P%','REB','AST','To', 'Pts','FG%']]
new_flb
```

```python
#Question 1: Average 3pt bl national team? 
avg = flb['3P%'].mean()

print('Question 1: What is the average pf 3pt in the national team? ')
print('The avg of 3Pts in the national team is ' , avg , '%')
```

```python
#Question 2: Best three 3pts shooter
top_3 = flb['3P%'].nlargest(3).index
top_3_players = flb.loc[top_3,['Player','Height','3P%']]
print('Question 2: Find the best three 3pts shooter')
print('Answer: ')
top_3_players
```

```python
#Question 3: Average 3pt la a3la 5 3pt shooters? 
top_5 = flb['3P%'].nlargest(5)
avg = top_5.mean()
print('Question 3: What is the average of 3pt of top 5 3pt shooters? ')
print('The avg of 3pts for top 5 3pts shooters is :' , avg , '%')
```

```python
#Question 4: for the top 5 3pts shooters, a) What is the avg of the height? b) What is the avg of FG?
print('Question 4: for the top 5 3pts shooters, a) What is the avg of the height? b) What is the avg of FG?')

top5 = flb['3P%'].nlargest(5).index
top5_players = flb.loc[top5]

avg1 = top5_players['Height'].mean()
print('a) The avg of the height of the top 5 3pts shooters is :' , avg1)

avg2 = top5_players['FG%'].mean()
print('b) The avg of FG% of the top 5 3pts shooters is :' , avg2 , '%')
```

```python
#Question 5: what player has the highest number of turnover?
flb = flb.head(18) 
most_TO = flb.loc[flb['To'].idxmax()]
player_with_most_TO = most_TO['Player']

print('Question 5: what player has the highest number of turnover?')
print('The player who has highest number of turnover is:', player_with_most_TO)
```

```python
#Question 6: What player do the most turnovers in a minute?
flb.loc[:,'TO_in_min'] = flb['To'] / flb['MIN']
flb = flb.head(18) 
most_TO_in_min = flb.loc[flb['TO_in_min'].idxmax()]
new = most_TO_in_min['Player']

print('Question 6: What player do the most turnovers in a minute?')
print('The player who has highest number of turnover per minutes is:' , new)

```

```python
#Question 7: If we take the top 5 tallest players, a) What is the avg of height ? b) Avg of 3pts? c) Avg of FG? 
height_5 = flb['Height'].nlargest(5).index
top_5height = flb.loc[height_5] 

print('Question 7: If we take the top 5 tallest players, a) What is the avg of height ? b) Avg of 3pts? c) Avg of FG? ')

avg2  = top_5height['Height'].mean()
print('a) The avg of height of the 5 tallest player is :' , avg2  , 'cm')

avg3 = top_5height['3P%'].mean()
print('b) The avg of 3pts of the 5 tallest player is :' , avg3  , '%')

avg4 = top_5height['FG%'].mean()
print('c) The avg of FG of the 5 tallest player is :' , avg4  , '%')

plt.bar(['avg of 3pts', 'avg of FG'],[ avg3 , avg4], color = ['green','black'])
plt.title('Average of height, 3pts and FG of the top 5 tallest players ')
plt.show()
```

```python
#Question 8: If we take top 3 assisters, what is the : a) Avg of assists b) Avg of height?
ast_3 = flb['AST'].nlargest(3).index
top3_ast = flb.loc[ast_3]

print('Question 8: If we take top 3 assisters, what is the : a) Avg of assists b) Avg of height?')

avg5 = top3_ast['AST'].mean()
print('a) The avg of assists of the top 3 assisters is: ' ,avg5, 'asts per game ')

avg6 = top3_ast['Height'].mean()
print('b) The avg of the height of the top 3 assisters is: ' ,avg6, 'cm ')
```
